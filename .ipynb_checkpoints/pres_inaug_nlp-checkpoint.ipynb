```python
def avalon_clean(avtext):
    import re
    soup = BeautifulSoup(avtext.text)
    
    doc = ''
    for para in soup.find_all('p'):
        doc += str(para)
    
    return re.sub('\<[^<>]*\>', ' ', doc)
```

```python
for avtext in [wa1, wa2, jeff1, jeff2, mad1, mad2, mon1, mon2, jqa,
              jack1, jack2, whh, polk, tay, pierce, linc1, linc2, grant1, grant2,
              hayes, cleve1, harr, cleve2, mckin1, mckin2, troos, taft, wils1, wils2,
              hard, hoov, fdr1, fdr2, fdr3, fdr4, trum, eisen1, eisen2, kenn,
              lbj, nix1, nix2, cart, reag1, reag2, ghwb, clin1, clin2, gwb1,
              gwb2, oba1]:
    corpus.append(avalon_clean(avtext))
```

```python
speeches = [nltk.tokenize.word_tokenize(speech) for speech in corpus]
```

```python
speeches
```

```python
speeches = [[word.lower() for word in speech] for speech in speeches]
```

```python
speeches
```

### Frequencies

```python
stops = stopwords.words('english')
stops.extend(list(string.punctuation))
```

```python
words = []
for speech in speeches:
    for word in speech:
        if word not in stops and not word.isnumeric():
            words.append(word)
```

```python
FreqDist(words)
```

```python
FreqDist(words).most_common()
```

## Exploration w/o Cross-Validation

```python
X_train, X_test, y_train, y_test = train_test_split(tfidfdf,
                                                   target,
                                                   stratify=target,
                                                   random_state=42)
```

```python
mnb = MultinomialNB()
mnb.fit(X_train, y_train)
mnb.score(X_train, y_train)
```

```python
mnb.score(X_test, y_test)
```

```python
mnb.predict(X_test)[mnb.predict(X_test)!='R']
```

```python
cnb = ComplementNB()
cnb.fit(X_train, y_train)
cnb.score(X_train, y_train)
```

```python
cnb.score(X_test, y_test)
```

```python
cnb.predict(X_test)[cnb.predict(X_test)!='R']
```

## Stacking

```python
sc = StackingClassifier(estimators=[
    ('rfc', RandomForestClassifier(bootstrap=False,
                            criterion='entropy',
                            max_features='log2',
                            n_estimators=1000,
                            random_state=42)),
    ('ertc', ExtraTreesClassifier(bootstrap=False,
                           criterion='gini',
                           max_features='sqrt',
                            n_estimators=1000,
                           random_state=42)),
    ('mnb', mnb),
    ('cnb', cnb)
])
```

```python
sc.fit(X_train, y_train)
```

```python
sc.score(X_test, y_test)
```

```python
sc.predict(X_test)
```

```python
sc.score(X_train, y_train)
```
